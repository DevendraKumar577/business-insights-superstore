```python
import pandas as pd

df = pd.read_csv("/content/Sample_ Superstore.csv")
df.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='mixed')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='mixed')

```

```python
df[['Order Date', 'Ship Date']].isna().sum()

```

```python
df[['Order Date', 'Ship Date']].dtypes
```

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Month Name'] = df['Order Date'].dt.month_name()
```

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
total_customers = df['Customer ID'].nunique()

total_sales, total_profit, total_orders, total_customers
```

```python
yearly_summary = (
    df.groupby('Order Year')[['Sales', 'Profit']]
    .sum()
    .reset_index()
)

yearly_summary
```

```python
monthly_sales = (
    df.groupby(['Order Year', 'Order Month Name'])['Sales']
    .sum()
    .reset_index()
)

monthly_sales.head()
```

```python
category_summary = (
    df.groupby('Category')[['Sales', 'Profit']]
    .sum()
    .reset_index()
)

subcategory_summary = (
    df.groupby('Sub-Category')[['Sales', 'Profit']]
    .sum()
    .reset_index()
)

category_summary, subcategory_summary.head()
```

```python
region_summary = (
    df.groupby('Region')[['Sales', 'Profit']]
    .sum()
    .reset_index()
)

region_summary
```

```python
df.to_csv('superstore_cleaned.csv', index=False)
yearly_summary.to_csv('yearly_summary.csv', index=False)
category_summary.to_csv('category_summary.csv', index=False)
region_summary.to_csv('region_summary.csv', index=False)
```
